```python
import pygame

# Initialize pygame
pygame.init()

# Set the window size
size = (300, 300)
screen = pygame.display.set_mode(size)

# Set the title of the window
pygame.display.set_caption("Tic-Tac-Toe")

# Load the X and O images
x_image = pygame.image.load("x.png")
o_image = pygame.image.load("o.png")

# Create a 2D list to represent the game board
board = [[0, 0, 0], [0, 0, 0], [0, 0, 0]]

# Keep track of whose turn it is (1 for X, -1 for O)
turn = 1

# Main game loop
running = True
while running:
    # Handle events
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

        # Handle mouse clicks
        elif event.type == pygame.MOUSEBUTTONDOWN:
            # Get the coordinates of the click
            x, y = event.pos
            # Calculate which square was clicked
            row = y // 100
            col = x // 100
            # Place the piece if the square is empty
            if board[row][col] == 0:
                board[row][col] = turn
                turn = -turn

    # Clear the screen
    screen.fill((255, 255, 255))

    # Draw the game board
    for r in range(3):
        for c in range(3):
            # Draw the appropriate piece (X or O)
            if board[r][c] == 1:
                screen.blit(x_image, (c * 100, r * 100))
            elif board[r][c] == -1:
                screen.blit(o_image, (c * 100, r * 100))

    # Update the screen
    pygame.display.flip()

# Quit pygame
pygame.quit()
```

```python
import tkinter as tk

class Calculator(tk.Tk):
    def __init__(self):
        super().__init__()

        self.title("Calculator")

        self.geometry("400x600")
        self.resizable(width=False, height=False)

        self.result = tk.Entry(self, width=40, borderwidth=5)
        self.result.grid(row=0, column=0, columnspan=4, padx=10, pady=10)

        self.create_button("1", 1, 1, 1)
        self.create_button("2", 1, 2, 1)
        self.create_button("3", 1, 3, 1)

        self.create_button("4", 2, 1, 1)
        self.create_button("5", 2, 2, 1)
        self.create_button("6", 2, 3, 1)

        self.create_button("7", 3, 1, 1)
        self.create_button("8", 3, 2, 1)
        self.create_button("9", 3, 3, 1)

        self.create_button("0", 4, 2, 1)
        self.create_button(".", 4, 3, 1)

        self.create_button("+", 5, 1, 1)
        self.create_button("-", 5, 2, 1)
        self.create_button("*", 5, 3, 1)
        self.create_button("/", 5, 4, 1)

        self.create_button("C", 6, 1, 1)
        self.create_button("=", 6, 2, 2)

    def create_button(self, text, row, column, columnspan):
        button = tk.Button(self, text=text, padx=40, pady=20, command=lambda: self.button_click(text))
        button.grid(row=row, column=column, columnspan=columnspan)

    def button_click(self, text):
        if text == "C":
            self.result.delete(0, tk.END)
        elif text == "=":
            try:
                self.result.delete(0, tk.END)
                self.result.insert(0, eval(self.expression))
            except:
                self.result.delete(0, tk.END)
                self.result.insert(0, "Error")
        else:
            current = self.result.get()
            self.result.delete(0, tk.END)
            self.result.insert(0, str(current) + str(text))
            self.expression = self.result.get()

app = Calculator()
app.mainloop()
```
